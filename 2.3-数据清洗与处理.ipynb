```python
data = data.dropna(subset=['房源特色'])
```

```python
data.shape
```

```python
# 2.3.4 数据预处理：均价int、面积int、去除无效字段
```

```python
data['均价'].head()
```

```python
data['均价'].str.split?
```

```python
data['均价'].str.split
```

```python
data['均价'] = data['均价'].str.split('元/平米',expand=True)[0]
```

```python
data['均价'].head()
```

```python
data['均价'] = data['均价'].astype('int')
```

```python
data['均价'].head()
```

```python
data['面积'] = data['面积'].str.split('平米',expand=True)[0]
```

```python
data['面积'] = data['面积'].astype('float')
```

```python
data.columns
```

```python
data = data[['链家编号','总价','均价','房型','面积',
             '小区名称', '所在区域','纬度', '经度']]
```
